```python
def get_num_videos(ator1, ator2):
    num = 0
    for x in videos_participants:
        atores_name = [a.name for a in x]
        if ator1 in atores_name and ator2 in atores_name:
            num += 1
    return num
```

```python
pornstars = []
tags = []
title = []
info = []
for x in range(len(videos_participants)):
    print(x)
    for y in range(len(videos_participants[x])):
        print(videos_participants[x][y].name)
        if videos_participants[x][y].name  == 'Abella-Anderson':
            continue
        try:
            index = pornstars.index(videos_participants[x][y].name)
            tags[index] = tags[index] + videos[x].tags
            title[index].append(videos[x].title)
            info[index].append(videos_participants[x][y].info)
        except ValueError:
            pornstars.append(videos_participants[x][y].name)
            tags.append(videos[x].tags)
            title.append([videos[x].title])
            info.append([videos_participants[x][y].info])
```

```python
tags = [[y.name for y in x] for x in tags]
```
